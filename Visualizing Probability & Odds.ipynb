# Visualizing Odds

$$odds = p / (1 - p)$$

## Create data covering the entire range of possible odds (p = 0 through 1)

Note: Since 1/0 returns a divide by zero error, I dropped p = 1.

```python
import numpy as np

import matplotlib.pyplot as pp
%matplotlib inline

p = np.arange(0,1, 0.01)

odds = p / (1 - p)
```

## Visualize the domain of possible oddds

### It Looks Exponential

Breaking the line into three regions, and thinking about what is going on in the odds equation, above, makes it easier to think about the relationship between probability and odds in an intuitive way. This is partially becuase in the context of the thrid section, the first two sections seem flat, so it is hard to appreciate what is going on if you look at the whole thing at once. But also breaking up the chart into three plots allows for a logical way to work out the way the balance in the equation changes accross the values of p.

```python
p1 = p[0:49]
odds1 = odds[0:49]

p2 = p[50:90]
odds2 = odds[50:90]

p3 = p[90:]
odds3 = odds[90:]

pp.plot(p1, odds1, 'blue')
pp.plot(p2, odds2, 'orange')
pp.plot(p3, odds3, 'green')
```

### From 0 to 0.5, the odds increase almost linearly, to 1

```python
pp.plot(p1, odds1, color = 'blue')
```

### When p = 0.75, the odds really start to take off

```python
pp.plot(p2, odds2, 'orange')
```

### p = >.89 is crazy growth, but looks like any other exponential curve without the context of p <= .89

The line is not smooth because I made data points every 0.01 units. And the slope is changing so rapidly, at this point, that you can see a big differance, even when I leave such small gaps.

```python
pp.plot(p3, odds3, 'green')
```

# Visualizing Probability of Odds

$$p = odds / (odds + 1)$$

I realize that I already have the probabilities, but re-calculating it to illustrate that the formula above actually works.

```python
p = odds / (odds + 1)

p1 = p[0:49]

p2 = p[50:90]

p3 = p[90:]

#pp.plot(odds, p)
pp.plot(odds1, p1, 'blue')
pp.plot(odds2, p2, 'orange')
pp.plot(odds3, p3, 'green')
```
